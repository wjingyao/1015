```python
import numpy as np
import pandas as pd
import seaborn as sb

import nltk
import string
from sklearn.feature_extraction.text import CountVectorizer
import pickle
import matplotlib.pyplot as plt

import gensim
from gensim import corpora
```

```python
data = pd.read_csv("reviews.csv")
description = data['comments'].astype(str)
```

```python
%run -i "contractions.py"

#Expanding Contractions
import re
contractions_re = re.compile('(%s)' % '|'.join(CONTRACTION_MAP.keys()))
def expand_contractions(s, CONTRACTION_MAP=CONTRACTION_MAP):
    def replace(match):
         return CONTRACTION_MAP[match.group(0)]
    return contractions_re.sub(replace, s)
```

```python
##Pre-Processing of Text 

#Lower case
description = description.apply(lambda x: x.lower())

#Expanding Contractions
description = description.apply(expand_contractions)

#remove punctuation
description = description.apply(lambda x : x.translate(str.maketrans('', '', string.punctuation)))

#remove numbers
description = description.apply(lambda x : x.translate(str.maketrans('', '', string.digits)))


#remove stop words
from nltk.corpus import stopwords
stop = stopwords.words('english')
description = description.apply(lambda x: " ".join(x for x in x.split() if x not in stop))
```

```python
#Lemmazation with POS tagging

from nltk.corpus import wordnet
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import word_tokenize, pos_tag
from collections import defaultdict
tag_map = defaultdict(lambda : wordnet.NOUN)
tag_map['J'] = wordnet.ADJ
tag_map['V'] = wordnet.VERB
tag_map['R'] = wordnet.ADV


lemma_function = WordNetLemmatizer()

def sentence_Lemmazation(sentence):
    tokens = word_tokenize(sentence)
    res_words = []
    for token, tag in pos_tag(tokens):
            res_words.append(lemma_function.lemmatize(token, tag_map[tag[0]]))
    return  " ".join(res_words)


descriptions = description.apply(sentence_Lemmazation)  
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence)))  

data = list(sent_to_words(description))

print(data[:1])
```

```python
dictionary = corpora.Dictionary(data)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in data]
```

```python
Lda = gensim.models.ldamodel.LdaModel
ldamodel = Lda(doc_term_matrix , num_topics = 3 , id2word= dictionary , passes = 50)
```

```python
print(ldamodel.print_topics(num_topics = 3 , num_words = 5))
```

```python
from gensim.models import CoherenceModel
coherence_model_lda = CoherenceModel(model=ldamodel, texts=data, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```

```python
def compute_coherence_values(dictionary, corpus, texts, limit, start=2, step=3):
    coherence_values = []
    model_list = []
    for num_topics in range(start, limit, step):
        Lda = gensim.models.ldamodel.LdaModel
        ldamodel = Lda(doc_term_matrix , num_topics = num_topics , id2word= dictionary , passes = 50)
        coherencemodel = CoherenceModel(model=ldamodel, texts=texts, dictionary=dictionary, coherence='c_v')
        coherence_values.append(coherencemodel.get_coherence())

    return model_list, coherence_values
```

```python
# Can take a long time to run.
model_list, coherence_values = compute_coherence_values(dictionary=dictionary, corpus=doc_term_matrix, texts=data, start=2, limit=40, step=6)
```

```python

limit=40; start=2; step=6;
x = range(start, limit, step)
for m, cv in zip(x, coherence_values):
    print("Num Topics =", m, " has Coherence Value of", round(cv, 4))
```
